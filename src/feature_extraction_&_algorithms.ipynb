```python
# Function to word tokenize
def function_word_tokenize(dataset, col_name):
  result = dataset[col_name].array

  for i in range(len(result)):
    result[i] = word_tokenize(str(result[i]), format='text')

  dataset[col_name] = result

# Handling
function_word_tokenize(dataset,'word_tokenize')
dataset[['content', 'word_tokenize']].head(40)
```

## Train_Test

```python
# Import libraries
from sklearn.model_selection import train_test_split

# Handling
data = dataset[['content', 'Sentiment']]
X = data['content']
y = data['Sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 6)
```

## Logistic Regression

```python
# Import libraries
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Handling
log_reg = Pipeline([('tf_idf', tfidf_vector),
                ('log_reg', LogisticRegression(random_state=0))])
```

```python
%%time
model_lr = log_reg.fit(X_train, y_train)
```

```python
model_lr.score(X_test, y_test)
```

```python
%%time
predict_lr = model_lr.predict(X_test)
```

```python
# Import libraries
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Result
accuracy_lr = accuracy_score(predict_lr, y_test)
print('Accuracy_score_lr: ',accuracy_lr)
print('Confusion_matrix_lr: \n', confusion_matrix(predict_lr, y_test))
print(classification_report(y_test, predict_lr))
```

## Support Vector Machine (SVM)

```python
# Import libraries
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

# Handling
svm = Pipeline([('tf_idf',tfidf_vector),
                ('svm', SVC(kernel='linear'))])
```

```python
%%time
model_svm = svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test, y_test)
```

```python
%%time
predict_svm = svm.predict(X_test)
```

```python
# Import libraries
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Result
accuracy_svm = accuracy_score(predict_svm, y_test)
print('Accuracy_score_svm: ',accuracy_svm)
print('Confusion_matrix_svm: \n', confusion_matrix(predict_svm, y_test))
print(classification_report(y_test, predict_svm))
```

## Naive Bayes

```python
# Import libraries
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

# Handling
naive_bayes = Pipeline([('tf_idf', tfidf_vector),
                        ('naive_bayes', MultinomialNB())])
```

```python
%%time
model_naive_bayes = naive_bayes.fit(X_train, y_train)
```

```python
model_naive_bayes.score(X_test, y_test)
```

```python
%%time
predict_naive_bayes = model_naive_bayes.predict(X_test)
```

```python
# Import libraries
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Result
accuracy_naive_bayes = accuracy_score(predict_naive_bayes, y_test)
print('Accuracy_score_naive_bayes: ',accuracy_naive_bayes)
print('Confusion_matrix_naive_bayes: \n', confusion_matrix(predict_naive_bayes, y_test))
print(classification_report(y_test, predict_naive_bayes))
```
