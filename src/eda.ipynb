```python
# Handling num_char
data_copy['num_char'] = ' '
for i in range(len(data_copy)):
    data_copy['num_char'][i] = len(str(data_copy['content'][i]))

# Handling num_word
data_copy['num_word'] = ''
for i in range(len(data_copy)):
    word_list = data_copy['content'][i].split(' ')
    data_copy['num_word'][i] = len(word_list)

# Handling avg_word_length
data_copy['avg_word_length'] = ''
for i in range(len(data_copy)):
    words = data_copy['content'][i].split(' ')
    average = sum(len(word) for word in words) / len(words)
    data_copy['avg_word_length'][i] = average

# Replace object -> float
data_copy['num_char'] = data_copy['num_char'].astype(float)
data_copy['num_word'] = data_copy['num_word'].astype(float)
data_copy['avg_word_length'] = data_copy['avg_word_length'].astype(float)


data_copy
```

Boxplot After removing outlier

```python
sns.set_style('darkgrid')
sns.set_palette('pastel')
plt.figure(figsize=(16,8))
df = pd.DataFrame(data = data_copy, columns = ['num_char','num_word','avg_word_length'])
boxplot = sns.boxplot(data=pd.melt(df), x = 'variable', y = 'value',  hue="variable")
boxplot.axes.set_title("Detect outliers", fontsize=24)
boxplot.set_xlabel("Variable", fontsize=18)
boxplot.set_ylabel("Value", fontsize=18)
plt.show()

# Box-plot before removing outliers
# plt.figure(figsize=(16,8))
# data_copy.boxplot(column=['num_char','num_word','avg_word_length'])
# plt.show()
```

```python
# def remove_outlier(col)
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
def check_outlier(lower, upper, data, col, col_check):
    for i in range(len(data)):
        if(data[col][i] < lower or data[col][i] > upper):
            data[col_check][i] = 'x'

    return data
```

```python
data_copy['check_outlier_num_char'] = ''
data_copy['check_outlier_num_word'] = ''
data_copy['check_outlier_avg_word_length'] = ''


# Drop the outlier value
# Replace the outlier value using the IQR


# num_char
low_num_char, up_num_char = remove_outlier(data_copy['num_char'])

# Start: Check outlier
data_copy =  check_outlier(low_num_char,up_num_char,data_copy,'num_char','check_outlier_num_char')
# End: Check outlier

data_copy['num_char'] = np.where(data_copy['num_char'] > up_num_char, up_num_char, data_copy['num_char'])
data_copy['num_char'] = np.where(data_copy['num_char'] < low_num_char, low_num_char, data_copy['num_char'])


# num_word
low_num_word, up_num_word = remove_outlier(data_copy['num_word'])

# Start: Check outlier
data_copy =  check_outlier(low_num_word,up_num_word,data_copy,'num_word','check_outlier_num_word')
# End: Check outlier

data_copy['num_word'] = np.where(data_copy['num_word'] > up_num_word, up_num_word, data_copy['num_word'])
data_copy['num_word'] = np.where(data_copy['num_word'] < low_num_word, low_num_word, data_copy['num_word'])


# avg_word_length
low_avg_word_length, up_avg_word_length = remove_outlier(data_copy['avg_word_length'])

# Start: Check outlier
data_copy =  check_outlier(low_avg_word_length,up_avg_word_length,data_copy,'avg_word_length','check_outlier_avg_word_length')
# End: Check outlier

data_copy['avg_word_length'] = np.where(data_copy['avg_word_length'] > up_avg_word_length, up_avg_word_length, data_copy['avg_word_length'])
data_copy['avg_word_length'] = np.where(data_copy['avg_word_length'] < low_avg_word_length, low_avg_word_length, data_copy['avg_word_length'])

data_copy.head()
data_copy.info()
```

```python
x= 0
y =0
z =0
for i in range(len(data_copy)):
    if(data_copy['check_outlier_num_char'][i] == 'x'):
        x+=1

    if(data_copy['check_outlier_num_word'][i] == 'x'):
        y+=1
        
    if(data_copy['check_outlier_avg_word_length'][i] == 'x'):
        z+=1

print("num_char: "+ str(x) + " d√≤ng")
print("num_word: "+ str(y) + " d√≤ng")
print("avg_word_length: "+ str(z) + " d√≤ng")
```
